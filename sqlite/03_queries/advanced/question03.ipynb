```python
%%sql
-- start with inner query
SELECT tr.AlbumId
FROM tracks tr
GROUP BY tr.AlbumId
HAVING COUNT(*) < 10
LIMIT 3;
```

```python
%%sql

SELECT ar.Name
FROM artists ar
JOIN albums al ON ar.ArtistId = al.ArtistId
WHERE al.AlbumId NOT IN (
    SELECT tr.AlbumId
    FROM tracks tr
    GROUP BY tr.AlbumId
    HAVING COUNT(*) < 10
)
GROUP BY ar.Name
HAVING COUNT(DISTINCT al.AlbumId) > 0; -- make sure the artist has at least one album
```

## Using `NOT EXISTS`

```python
%%sql 

SELECT ar.Name
FROM artists ar
JOIN albums al ON ar.ArtistId = al.ArtistId
WHERE NOT EXISTS (
    SELECT 1
    FROM tracks tr
    WHERE tr.AlbumId = al.AlbumId
    GROUP BY tr.AlbumId
    HAVING COUNT(*) < 10
)
GROUP BY ar.Name
HAVING COUNT(DISTINCT al.AlbumId) >0
```
