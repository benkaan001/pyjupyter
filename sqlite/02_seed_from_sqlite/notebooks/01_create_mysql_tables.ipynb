```python
import sqlite3
import mysql.connector
import sys 
sys.path.append('..') # ensure utils is reachable
from utils.config import MYSQL_CONFIG
```

```python
# set up SQLite connection
with sqlite3.connect('../databases/MediaStore.db') as sqlite_conn:
    sqlite_cursor = sqlite_conn.cursor()

    # retrieve table names from sqlite_master table
    sqlite_cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    table_names = [result[0] for result in sqlite_cursor.fetchall()]

    # set up MySQL connection
    with mysql.connector.connect(**MYSQL_CONFIG) as mysql_conn:
        mysql_cursor = mysql_conn.cursor()

        # drop existing tables in MySQL
        for table_name in table_names:
            mysql_cursor.execute(f"DROP TABLE IF EXISTS {table_name};")
            print(f"Table {table_name} dropped from MySQL")

        # create new tables in MySQL
        with open('../databases/MediaStore.sql') as f:
            sql_script = f.read()
            print(f"    {sql_script}")

        mysql_cursor.execute(sql_script)
        print("Tables created in MySQL")
```
