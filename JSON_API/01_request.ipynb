
- This is a listing of all packages available from the core tap via the Homebrew package manager for macOS and Linux.
  - `https://formulae.brew.sh/api/formula.json` 

- Sample analytics API url for node package.
  - Formula JSON API: `/api/formula/node.json`



```python
import requests

# url containing metadata about all packages
url = "https://formulae.brew.sh/api/formula.json"

r = requests.get(url)

# using the json decoder of the requests module, read the JSON response 
packages_json = r.json()

# print the first package object in the raw json response list
sample_package_json = packages_json[0]
print(sample_package_json)
```

```python
# using the json module, dump the response to a string, so it can be formatted 
import json

# sample package object string 
sample_package_str = json.dumps(sample_package_json, indent=2)

print(sample_package_str)
```

```python
import json

# sample analytics url for a given package
package_name = "node"
analytics_url = f"https://formulae.brew.sh/api/formula/{package_name}.json"

r = requests.get(analytics_url)

# using the json decoder of the requests module, read the JSON response 
node_package_json = r.json()


# json string
node_package_str = json.dumps(node_package_json, indent=2)

print(node_package_str)
```
