```python
# print out the top 10 packages that has video in its description based on the 
# 365d install_on_request value in DESC order. 

import json 

with open("package_info.json", 'r') as f:
    data = json.load(f)

# filter out the packages in the data that has AWS in its description 
data = [package for package in data if 'AWS' in package['desc']]

sorted_data = sorted(data, key= lambda x: x['analytics']['365d'], reverse=True)[0:10]

sorted_data_str = json.dumps(sorted_data, indent=2)

print(sorted_data_str)
```
