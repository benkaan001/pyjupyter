```python
from csv import DictReader
from path import survey_path
from collections import Counter


# initiate a dictionary to hold devtype info
devtype_info = {}

with open(survey_path) as f:
    csv_reader = DictReader(f)


    for row in csv_reader:
        devtypes = row['DevType'].split(";")

        for devtype in devtypes:
            # initiate the value to an empty dict
            devtype_info.setdefault(devtype, {
                                            'total': 0,
                                            'language_counter': Counter()
                                            }
            ) 
            languages = row['LanguageHaveWorkedWith'].split(";")
            devtype_info[devtype]['language_counter'].update(languages)
            devtype_info[devtype]['total'] += 1
            

for dev_type, info in devtype_info.items():
    print(f"{dev_type}")

    for language, value in info['language_counter'].most_common(5):
        lang_percentage = round ( (value/info['total']) * 100, 2)
        print(f"\t{language}: {lang_percentage}%")
        
```
